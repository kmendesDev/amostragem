```python
df = pd.DataFrame({"Grupo":b,"Valores":a})
df
```

##Criando função para extrair uma quantidade de termos conforme o grupo

```python
def amostra_estratificada(df,n,estrato):
  amostra = df.groupby(estrato, group_keys=False).apply(lambda x: x.sample(min(len(x),n)))

  return amostra
```

##Estratos com tamanho 2:

```python
amostra_estratificada(df,2,'Grupo')
```

```python
amostra_estratificada(df,3,'Grupo')
```

## Função de para extrair dados conforme o tamanho da amostra final

```python
def amostra_estrat_2(df, N, estrato):
  tam_pop = df.shape[0]
  amostra = df.groupby(estrato, group_keys=False)\
    .apply(lambda x: x.sample(int(N*len(x)/tam_pop)))\
    .reset_index(drop=True)\
    .sort_values(by=estrato)
  return amostra
```

##Amostra de tamanho 9:

```python
amostra_estrat_2(df,9,'Grupo')
```

#Amostragem por Conglomerado:

```python
def amostra_conglomerado(df, n_cong, cong):
  todos_cong = list(df[cong].unique())
  tam_cong = len(todos_cong)
  n = min(n_cong, tam_cong)
  cong_sorteados = sample(todos_cong, n)

  amostra = df[df[cong].isin(cong_sorteados)]
  return amostra
```

##Escolhendo 2 conglomerados

```python
amostra_conglomerado(df,2,'Grupo')
```

##Escolhendo 3 conglomerados

```python
amostra_conglomerado(df,3,'Grupo')
```
